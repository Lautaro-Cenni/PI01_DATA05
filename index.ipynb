```python
import pandas as pd

df_amazon = pd.read_csv('Datasets/amazon_prime_titles.csv')
#separo las columnas que voy a utilizar en cada dataframe
df_amazon_columns = df_amazon[['title','type','release_year','cast','duration','listed_in']]
df_amazon_columns['plataforma'] = 'Amazon'
df_amazon_columns
```

```python
df_disneyplus = pd.read_csv('./Datasets/disney_plus_titles.csv')

df_disneyplus_columns = df_disneyplus[['title','type','release_year','cast','duration','listed_in']]
df_disneyplus_columns['plataforma'] = 'Disney Plus'
df_disneyplus_columns
```

```python
df_hulu = pd.read_csv('./Datasets/hulu_titles.csv')

df_hulu_columns = df_hulu[['title','type','release_year','cast','duration','listed_in']]
df_hulu_columns['plataforma'] = 'Hulu'
```

```python
df_netflix = pd.read_json('./Datasets/netflix_titles.json')

df_netflix_columns = df_netflix[['title','type','release_year','cast','duration','listed_in']]
df_netflix_columns['plataforma'] = 'Netflix'
```

```python
#concateno todos los dataframes en una sola tabla
df_unificado = pd.concat([df_amazon_columns,df_disneyplus_columns,df_hulu_columns,df_netflix_columns],ignore_index=True)

#creo una columna para los indicies
df_unificado['indice'] = df_unificado.index

#verifico las columnas que tienen nulos
df_unificado.isnull().sum()
```

```python
#reemplazamos los valores nulos en la columna cast por 'sin dato' porque no tengo forma de rellenar una cantidad significativa de datos en esta columna
df_unificado['cast'].fillna('sin dato',inplace=True)

#decidi cambiar los nombres para que sean mas intuitivos y en español
column_names = {'title':'titulo','type' : 'clasificacion', 'release_year' : 'fecha_de_estreno','cast' : 'elenco', 'duration' : 'duracion', 'listed_in' : 'lista_de_generos'}
df_unificado.rename(columns=column_names,inplace=True)
```

```python
#voy a analizar la columna duracion para ver si puedo hacer algo para llenar los espacio vacios
#tengo este dato como "cantidad de tiempo" y "minutos o season" asi que las separo en dos, una con numeros y otra con string
df_duracion = df_unificado["duracion"].str.split(expand=True)
df_unificado['duracion_cantidad']=df_duracion[0]
df_unificado['duracion_minOSeason']=df_duracion[1]
```

```python
#cambio el orden de las columnas para que sea mas comodo
df_unificado = df_unificado[['indice','titulo','clasificacion','fecha_de_estreno','elenco','duracion','duracion_cantidad','duracion_minOSeason','lista_de_generos','plataforma']]
```

```python
df_unificado.head()
```

```python
# verifico todos los maximos para ver si me conviene reemplazar valores por media,mediana,etc
max_lists = df_unificado.loc[df_unificado['duracion_minOSeason'] == 'min'][['plataforma','fecha_de_estreno','duracion_cantidad']]
#primero tome los valores que son estan en minutos, si la unidad es min entonces no es nulo el dato
max_lists['duracion_cantidad'] = max_lists['duracion_cantidad'].astype('int64')
max_lists = max_lists.groupby(['plataforma','fecha_de_estreno'])['duracion_cantidad'].max()
print(max_lists['Amazon'].sort_values()) #este es un ejemplo
#hay numeros muy variados de datos por lo que sacar un promedio no seria representativo, prefiero dejarlo como 'sin dato' para que al menos figure como un dato a cambiar en el futuro en el hipotetico caso en el que se revisaran los datos.
#ademas de que no aportan nada para las consultas que piden los enunciados
df_unificado.fillna('sin dato',inplace=True)
```
